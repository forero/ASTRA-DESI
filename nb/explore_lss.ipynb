```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
# original files are found here https://data.desi.lbl.gov/public/edr/vac/edr/lss/v2.0/LSScats/clustering/
lss_data = Table.read("../data/edr/vac/edr/lss/v2.0/LSScats/clustering/BGS_ANY_S_clustering.dat.fits")
lss_rand = Table.read("../data/edr/vac/edr/lss/v2.0/LSScats/clustering/BGS_ANY_S_17_clustering.ran.fits")
```

```python
set(lss_data['ROSETTE_NUMBER'])
```

```python
ii = lss_data['ROSETTE_NUMBER']==10
plt.scatter(lss_data['RA'][ii], lss_data['DEC'][ii], s=0.1)
plt.axis('equal')  
```

```python
_=plt.hist(lss_data['Z'][ii])

kk = lss_rand['ROSETTE_NUMBER']==10
_=plt.hist(lss_rand['Z'][kk], alpha=0.4)
```

```python
jj = (lss_data['Z']>0.3) & (lss_data['Z']<0.32)
plt.scatter(lss_data['RA'][ii&jj], lss_data['DEC'][ii&jj], s=5)
plt.axis('equal')  
```

```python
kk = lss_rand['ROSETTE_NUMBER']==10
ll = (lss_rand['Z']>0.3) & (lss_rand['Z']<0.32)
plt.scatter(lss_rand['RA'][kk&ll], lss_rand['DEC'][kk&ll], s=5)
plt.scatter(lss_data['RA'][ii&jj], lss_data['DEC'][ii&jj], s=5)
plt.axis('equal')  
```
